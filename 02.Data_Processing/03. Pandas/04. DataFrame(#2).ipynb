## Boolean, Arithmetics

```python
df1 = pd.DataFrame(np.random.randint(4, 20, size=(4, 3)), columns=["a", "b", "c"], dtype="int64")
df2 = pd.DataFrame(np.random.randint(4, 20, size=(3, 4)), columns=["a", "b", "c", "d"], dtype="int64")
```

```python
df1 + df2
```

```python
df1 - df2.iloc[2]
```

```python
1 / df1
```

```python
df2 ** 4
```

```python
df1 = pd.DataFrame({"a": [0, 10, 5], "b": [6, 7, 0]})
df2 = pd.DataFrame({"a": [12, 3, 1], "b": [19, 4, 55]})
```

```python
df1 | df2
```

```python
np.exp(df2)
```

```python
np.asarray(df2)
```

```python
np.remainder(df1, df2)
```
